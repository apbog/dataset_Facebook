## Часть первая.
Вычисление среднего, максимального/минимального значений, медианы, моды числовых значений для всего массива в целом и для каждого типа контента (столбец Type) в отдельности.

***Загрузка данных из массива в DataFrame и вывод первых 10 объектов.***

```python
data = pd.read_csv('dataset_Facebook.csv', sep = ';')
data.head(10)
```

***Список всех типов контента***

```python
data.Type.unique()
```

***Создаем новый DataFrame строки которого показывают среднее(Mean), максимальное(Max), минимальное(Min), медиану(Med) и моду(Mode) числовых значений массива.
Столбец 'Type' не выводится, поскольку не является числовым.***

```python
df = data
lst = df.columns.tolist()
lst.remove('Type')
ans = pd.DataFrame({'Mean': df.mean(), 'Max': df.max(), 'Min': df.min(), 'Med': df.quantile(q=0.5), 'Mode': df.mode().iloc[0]}, index = df.columns)
ans = ans.T
ans.loc[['Mean', 'Max', 'Min', 'Med', 'Mode'], lst]
```

## Часть вторая.
Найти самый популярный объект в выборке, объяснить почему

***Для расчета популярности обьектов (постов) был выбран метод с коэффициентом вовлеченности в пересчете на охват (engagement rate by reach или ERR).
ERR = (Total reactions)/(Reach) ⋅ 100% 
или же в переводе на столбцы матрицы:***

***ERR = Total Interactions / Lifetime Post Total Reach ⋅ 100%***

*** Итоговое значение ERR показывает % людей, которые проявили активность у конкретного поста.***

*** В данном методе не используется информация о людях, которым понравилась страница с постом (подписчиках)[Lifetime Post reach by people who like your Page] и их активности на посте[Lifetime People who have liked your Page and engaged with your post].
Это связано с тем, что коеффицент ERR расчитаный на основании только подписчиков сильно превышает общий ERR и не всегда является оптимальным показателем (Пример - обьект № 118 - ERR_by_like = 49.86% ERR = 1.08% ;
обьект № 87 - ERR_by_like = 18.17% ERR = 7.75%). 
Исходя из приведенного примера видно, что коеффицент ERR_by_like целесообразно использовать в задачах, в которых вес активности подписчиков группы превышает вес активности новых (сторонних) пользователей.***

***Для более точного анализа необходимо понимать, какие критерии весят больше прочих. 
Т.е. для разных показателей поста добавляются множители. 
Например, вес репоста больше веса комментария, а вес комментария в свою очередь больше веса лайка, тогда можно выразить 
ER как ***

*** (share ⋅ 2 + comment + like / 2) / Lifetime Post Total Reach ⋅ 100%***

*** Кроме того, можно добавить такие показатели как: вовлеченность у подписчиков, общее количество показов поста(Lifetime Post Total Impressions), общее число подписчиков страницы и другие. ***

```python
lst = [0,1,7,8,9,12,13,14,18,19]
data['ERR'] = data['Total Interactions'] / data['Lifetime Post Total Reach'] * 100
data.iloc[((data['ERR'].sort_values(ascending=False)).head(5)).keys().tolist(),lst]
```

*** По оценке коэффицента ERR самым популярным (по % людей проявивших активность к посту) является пост № 87. Из 4012 людей просмотревших данный пост 311 (7.75%) проявили активность.***

*** Сравнение популярности постов у подписчиков страницы (ERR_F) и всех пользователей (ERR)***

```python
lst.append(20)
data['ERR_F'] = data['Lifetime People who have liked your Page and engaged with your post'] / data['Lifetime Post reach by people who like your Page'] * 100
data.iloc[((data['ERR_F'].sort_values(ascending=False)).head(3)).keys().tolist(),lst]
```

*** В таблице ниже приведены три наиболее просматриваемых по числу пользователей поста. Однако показатели ERR равняются 3.5%, 0.2% и 0,11% соответственно. Что говорит о плохой активности пользователей и как следствие - низкой популярности. ***

*** Активность самого просматриваемого поста (№244) в 2.2 раза меньше активности поста №87(ERR = 7.75)***

```python
data.iloc[((data['Lifetime Post Total Reach'].sort_values(ascending=False)).head(3)).keys().tolist(),lst]
```

*** В таблице ниже представлены три поста с наибольшим колличеством активных действий. ERR равняется 3.5%, 2.06% и 4.7% соответственно.***

*** При этом пост с наибольшим колличеством активных действий также является наиболее просматриваемым постом. ***

```python
data.iloc[((data['Total Interactions'].sort_values(ascending=False)).head(3)).keys().tolist(),lst]
```

*** Пост с наибольшим колличеством просмотров, среди людей, которым понравилась страница (Подписчиков) - № 244 ***

```python
data.iloc[((data['Lifetime Post reach by people who like your Page'].sort_values(ascending=False)).head(1)).keys().tolist(),lst]
```

## Итог
*** В результате можно заключить, что самым популярным постом по коэффициенту вовлеченности в пересчете на охват является пост под номером 87 (ERR = 7.75, просмотревших людей = 4012, активных действий = 311). ***

*** В то время как самым популярным с точки зрения количества людей, просмотревших пост (180480 человек), количества подписчиков, просмотревших пост(51456 человек), и количества активных действий у поста (372 комментария, 5172 лайка и 790 репостов) является пост под номером 244 (ERR = 3.5) ***

```python
print("Обьект №87:" + '\n' + str(data.iloc[87]))
print('\n' + "Обьект №244:" + '\n' + str(data.iloc[244]))
```
